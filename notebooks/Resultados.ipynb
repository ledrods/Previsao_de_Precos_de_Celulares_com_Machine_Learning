## Impacto da Limpeza de Dados na Matriz de Correlao

A limpeza de dados foi fundamental para aprimorar a preciso da anlise, especialmente na **matriz de correlao**. Inicialmente, ao comparar o **preo dos celulares dos EUA** com a mdia do dataset, a correlao era de **0.28**, o que indicava uma relao fraca e distorcida. Porm, aps a **limpeza de dados**, essa correlao subiu para **0.99**, refletindo uma relao muito mais robusta e justa.

## O Outlier e a Distorso da Mdia

Esse aumento na correlao se deve a um nico **outlier** que estava **distorcendo a mdia** dos preos dos EUA. Esse valor extremamente alto fazia com que a mdia fosse **inflacionada**, gerando uma comparao injusta com outros pases.

### Investigao do Outlier

Busquei investigar a origem desse valor discrepante, considerando a possibilidade de que fosse um modelo de celular de **edio limitada** com um preo mais elevado. Contudo, ao analisar mais detalhadamente, percebi que o valor elevado no se referia a uma edio especial, mas sim a um **erro grosseiro no dataset**, onde um celular comum estava sendo listado a um preo exorbitante.

## Converso de Moedas e Impacto na Anlise

Outro ponto crucial foi a **converso dos preos para USD**, garantindo que a anlise fosse consistente entre todas as moedas. Cada preo foi **convertido com base nas taxas de cmbio do perodo**, permitindo comparaes justas entre os valores de lanamento dos celulares em diferentes pases. Aps essa converso, foi possvel perceber que o **Paquisto teve a menor correlao com a mdia de preos, atingindo um coeficiente de 0.92**. Esse dado sugere que os preos de lanamento no Paquisto apresentam **maior variao em relao ao valor mdio global**.

## Resultado da Limpeza de Dados

A remoo do **outlier** e a **converso das moedas** tiveram um **impacto direto na anlise**, tornando a **mdia dos preos dos EUA mais representativa** e permitindo uma **comparao mais precisa** entre os preos dos celulares em diferentes pases.
