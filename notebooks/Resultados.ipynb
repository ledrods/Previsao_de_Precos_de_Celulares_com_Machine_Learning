# Impacto da Limpeza de Dados na Matriz de Correlao

A limpeza de dados foi fundamental para aprimorar a preciso da anlise, especialmente na **matriz de correlao**. Inicialmente, ao comparar o **preo dos celulares dos EUA** com a mdia do dataset, a correlao era de **0.28**, o que indicava uma relao fraca e distorcida. Porm, aps a **limpeza de dados**, essa correlao subiu para **0.93**, refletindo uma relao muito mais robusta e justa.

## O Outlier e a Distorso da Mdia

Esse aumento na correlao se deve a um nico **outlier** que estava **distorcendo a mdia** dos preos dos EUA. Esse valor extremamente alto fazia com que a mdia fosse **inflacionada**, gerando uma comparao injusta com outros pases.

### Investigao do Outlier

Busquei investigar a origem desse valor discrepante, considerando a possibilidade de que fosse um modelo de celular de **edio limitada** com um preo mais elevado. Contudo, ao analisar mais detalhadamente, percebi que o valor elevado no se referia a uma edio especial, mas sim a um **erro grosseiro no dataset**, onde um celular comum estava sendo listado a um preo exorbitante.

## Resultado da Limpeza de Dados

A remoo desse erro teve um **impacto direto na anlise**, tornando a mdia dos preos dos EUA mais representativa e permitindo uma comparao **mais precisa** entre os preos dos celulares em diferentes pases.
